```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# The questions
questions = ["What is your program?",
            "How satisfied are you with the teaching quality of your program",
            "How satisfied are you with the facilities in your program?",
            "How satisfied are you with the materials available to you in your program?",
            "How likely would you recommend your program to other?"
           ]

# The responses of the users will be based on a scale of 1-10

# Programs
def take_survey(questions):
    responses = {}
    i = 0
    while i < len(questions):
        question = questions[i]
        response = ''
        while response == '':
            response = input(question + " ")
        responses[question] = response
        i +=1
    return responses

# Store input inside a DataFrame
survey_data = pd.DataFrame(columns= questions)

#Ask user if they are interested in the survey
while True:
    take_survey_input = input('Do you want to take the survey? (y/n)')
    if take_survey_input == 'y':
        responses = take_survey(questions)
    
        
        #Store (Append) the user input
        survey_data = survey_data.append(responses, ignore_index= True)
    else:
        break
        
#Visuals
#Satisfaction Level
satisfaction_columns = ["How satisfied are you with the teaching quality of your program",
            "How satisfied are you with the facilities in your program?",
            "How satisfied are you with the materials available to you in your program?",
            "How likely would you recommend your program to other?"
           ]
satisfaction_avg = survey_data[satisfaction_columns].astype(int).mean()

# Bar Chart
plt.bar(satisfaction_avg.index, satisfaction_avg.values)
plt.title("Satisfaction Average for each questions")
plt.xlabel("Questions")
plt.ylabel("Average Range")
plt.xticks(rotation=90)
plt.show()

#Print DataFrame
survey_data
```

```python
#Pie Chart
plt.pie(satisfaction_avg.values, labels= satisfaction_avg.index)
plt.title("Satisfaction Average for each questions")
plt.show()
```

```python
# Horizontal Bar Chart
plt.barh(satisfaction_avg.index, satisfaction_avg.values)
plt.title("Satisfaction Average for each questions")
plt.xlabel("Questions")
plt.ylabel("Average Range")
plt.show()
```

```python
# Line Chart
plt.plot(satisfaction_avg.index, satisfaction_avg.values, marker = 'o')
plt.title("Satisfaction Average for each questions")
plt.xlabel("Questions")
plt.ylabel("Average Range")
plt.xticks(rotation=90)
plt.show()
```

```python
# Histogram
plt.hist(survey_data["How likely would you recommend your program to other?"])
plt.title("Distribution of Programs based on recommendation")
plt.xlabel("Range")
plt.ylabel("e-Range")
plt.show()
```
